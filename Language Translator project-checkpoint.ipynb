```python
# Import necessary libraries
from tkinter import *
from tkinter import ttk
from googletrans import Translator

def Translate():
    # Disable the Translate button during translation
    trans_btn.config(state=DISABLED)

    # Create a Translator object
    translator = Translator()

    # Translate the input text to the selected destination language
    translated = translator.translate(text=Input_text.get(), dest=dest_lang.get())

    # Clear the output text box and insert the translated text
    Output_text.delete(1.0, END)
    Output_text.insert(END, translated.text)

    # Enable the Translate button after translation
    trans_btn.config(state=NORMAL)

# Create the main window
root = Tk()
root.geometry("1200x420")  # Set the initial window size
root.resizable(0, 0)       # Disable window resizing
root.configure(bg="blue")  # Set the background color
root.title("Language Translator by JK_batch_05")  # Set the window title

# Label for the application title
Label(root, text="Language Translator", font="Arial 20 bold", bg="yellow").pack()

# Label for input
Label(root, text="Enter Your Text", font="Arial 13 bold", bg="yellow").place(x=160, y=100)

# Input text box
Input_text = Entry(root, font="Arial 12", width=50)
Input_text.place(x=300, y=100)

# Label for output
Label(root, text="Output", font="Arial 13 bold", bg="yellow").place(x=190, y=150)

# Label for credit
Label(root, text="Developed by: Jawad Khan", font="Arial 20 bold", bg="red").place(x=805, y=310)

# Output text box
Output_text = Text(root, font="Arial 10", height=5, wrap=WORD, padx=5, width=50)
Output_text.place(x=300, y=130)

# Dropdown for selecting destination language
languages = [
    "Afrikaans", "Albanian", "Amharic", "Arabic", "Armenian", "Azerbaijani",
    "Basque", "Belarusian", "Bengali", "Bosnian", "Bulgarian", "Catalan",
    "Cebuano", "Chichewa", "Chinese (Simplified)", "Chinese (Traditional)",
    "Corsican", "Croatian", "Czech", "Danish", "Dutch", "English", "Esperanto",
    "Estonian", "Filipino", "Finnish", "French", "Urdu"
]

dest_lang = ttk.Combobox(root, values=languages, width=22)
dest_lang.place(x=800, y=100)
dest_lang.set("select launguage")  # Set the initial selection

# Translate button
trans_btn = Button(root, text="Translate", font="Arial 12 bold", pady=5, command=Translate, bg="red")
trans_btn.place(x=680, y=150)

# Start the Tkinter main loop to run the GUI application
root.mainloop()
```
